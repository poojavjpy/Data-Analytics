# ***7. Create simple visualizations including pie charts, bar charts, and histograms using Seaborn.***

```python
import pandas as pd

# Simple dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math": [85, 70, 90, 60, 75],
    "Science": [80, 65, 88, 72, 70],
    "English": [78, 82, 85, 68, 74]
}

df = pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Bar Chart (Average marks of each subject) ---
avg_marks = df[["Math", "Science", "English"]].mean()
sns.barplot(x=avg_marks.index, y=avg_marks.values)
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.show()

# --- 2. Histogram (Distribution of Math marks) ---
sns.histplot(df["Math"], bins=5, kde=True)
plt.title("Distribution of Math Marks")
plt.xlabel("Marks")
plt.show()

# --- 3. Pie Chart (Marks of one student, e.g., Student A) ---
marks_A = df[df["Student"] == "A"].iloc[0, 1:]  # Student A's marks
plt.pie(marks_A, labels=marks_A.index, autopct="%1.1f%%")
plt.title("Student A's Marks Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dataset
data = {
    "Class": ["Class 1", "Class 2", "Class 3", "Class 4"],
    "Students": [40, 35, 50, 45]
}
df = pd.DataFrame(data)

# Bar chart
sns.barplot(x="Class", y="Students", data=df)
plt.title("Number of Students in Each Class")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dataset
marks = {"Marks": [45, 55, 60, 65, 70, 75, 80, 82, 85, 90]}
df = pd.DataFrame(marks)

# Histogram
sns.histplot(df["Marks"], bins=5, kde=True)
plt.title("Distribution of Exam Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataset
sales = {"Product": ["Apples", "Bananas", "Cherries", "Dates"],
         "Sales": [50, 30, 10, 10]}

# Pie chart
plt.pie(sales["Sales"], labels=sales["Product"], autopct="%1.1f%%", startangle=90)
plt.title("Sales Distribution by Product")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Bigger dataset
data = {
    "Department": ["HR", "HR", "IT", "IT", "IT", "Sales", "Sales", "Sales", "Finance", "Finance"],
    "Salary": [40000, 42000, 60000, 65000, 70000, 50000, 52000, 48000, 55000, 53000]
}
df = pd.DataFrame(data)

# Bar chart of average salary by department
avg_salary = df.groupby("Department")["Salary"].mean().reset_index()

sns.barplot(x="Department", y="Salary", data=avg_salary)
plt.title("Average Salary by Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Bigger dataset (simulated house prices)
np.random.seed(42)
house_prices = np.random.normal(500000, 80000, 200)  # mean=5L, std=80k, 200 houses

df = pd.DataFrame({"Price": house_prices})

# Histogram
sns.histplot(df["Price"], bins=15, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("House Price ()")
plt.ylabel("Number of Houses")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataset
smartphones = {
    "Brand": ["Samsung", "Apple", "Xiaomi", "Oppo", "Vivo", "Others"],
    "Market_Share": [270, 230, 200, 150, 120, 180]
}

# Pie chart
plt.pie(
    smartphones["Market_Share"],
    labels=smartphones["Brand"],
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Global Smartphone Market Share")
plt.show()
```
