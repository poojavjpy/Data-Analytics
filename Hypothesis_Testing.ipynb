üéØ Problem Statement (Example)

A company claims that the average battery life is 500 hours.
A sample of 30 batteries has:

Sample mean = 480 hours

Sample standard deviation = 50 hours

We will:

Perform a one-sample t-test

Compute a 95% confidence interval

Draw a conclusion

üìå Step 1: Import Required Libraries

```python
import numpy as np
from scipy import stats
```

üìå Step 2: Given Sample Information

```python
# Given data
sample_mean = 480
population_mean = 500
sample_std = 50
n = 30
alpha = 0.05  # 5% significance level
```

üìå Step 3: Hypothesis Testing (One-Sample t-Test)
Hypotheses:

H‚ÇÄ (Null Hypothesis): Mean battery life = 500 hours

H‚ÇÅ (Alternative Hypothesis): Mean battery life ‚â† 500 hours

```python
# Calculate t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Critical t-value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

print("T-statistic:", round(t_statistic, 2))
print("Critical t-value:", round(t_critical, 2))
```

üìå Step 4: Decision Rule

```python
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

üìå Step 5: 95% Confidence Interval

```python
margin_of_error = t_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("95% Confidence Interval:")
print("(", round(lower_bound, 2), ",", round(upper_bound, 2), ")")
```

```python
print("T-statistic:", -2.19)
print("Critical t-value:", 2.05)

print("Reject the null hypothesis")
print("95% Confidence Interval:")
print("(461.34 , 498.66)")
```

üß† Interpretation (Very Important for Students)

The claimed mean (500) is not inside the confidence interval

The t-statistic exceeds the critical value
‚úîÔ∏è Therefore, the company's claim is not valid
